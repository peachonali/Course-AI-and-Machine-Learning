#1 สร้างเกมปริศนาเลื่อน 8 Puzzle

```python
from collections import deque

def bfs(start_state):
    target = [1, 2, 3, 4, 5, 6, 7, 8, 0]
    dq = deque([start_state])
    visited = {tuple(start_state): None}

    while dq:
        state = dq.popleft()
        if state == target:
            path = []
            while state :
                path.append(state)
                state = visited[tuple(state)]
            return path[::-1]

        zero = state.index(0)
        row, col = divmod(zero, 3)
        for move in (-3, 3, -1, 1):
            new_row,new_col = divmod(zero + move,3)
            if 0 <= new_row < 3 and 0 <= new_col < 3 and abs(row - new_row) + abs(col - new_col) == 1:
                  neighbor = state[:]
                  neighbor[zero], neighbor[zero + move] = neighbor[zero + move], neighbor[zero]
                  if tuple(neighbor) not in visited:
                      visited[tuple(neighbor)] = state
                      dq.append(neighbor)

def printSolution(path):
    for state in path:
        print("\n".join(' '.join(map(str, state[i:i+3])) for i in range(0, 9, 3)))
        print("-----")

# Example Usage
startState = [1, 3, 0, 6, 8, 4, 7, 5, 2]
solution = bfs(startState)

if solution:
    printSolution(solution)
    print(f"Solved in {len(solution) - 1} moves.")
else:
    print("No solution found.")
```

```python
from collections import deque              # import deque เพื่อใช้เป็นคิว (queue) สำหรับ BFS ซึ่ง pop ซ้ายได้เร็ว
#.ใช้แทน list เพราะ หาค่าทางซ้ายได้เร็วกว่า o(1) list o(n)
def bfs(start_state):                                  # ฟังก์ชัน bfs รับสถานะเริ่มต้นของบอร์ด 8-puzzle
    target = [1, 2, 3, 4, 5, 6, 7, 8, 0]                # กำหนดสถานะเป้าหมาย (goal state) index 8 , 9 เลข
    dq = deque([start_state])                          # สร้างคิว BFS และใส่สถานะเริ่มต้นลงไป
    visited = {tuple(start_state): None}               # เก็บสถานะที่เคยไปแล้ว โดย value คือ parent ของ state นั้น

    while dq:                                          # วนลูปตราบใดที่คิวยังไม่ว่าง
        state = dq.popleft()                           # ดึงสถานะแรกออกจากคิว (FIFO → BFS)
        if state == target:                            # ตรวจสอบว่าถึงสถานะเป้าหมายหรือยัง
            path = []                                  # สร้าง list สำหรับเก็บเส้นทางคำตอบ
            while state:                               # วนย้อนกลับจาก goal ไป start โดยอาศัย visited
                path.append(state)                     # เก็บ state ปัจจุบันลงใน path
                state = visited[tuple(state)]          # เปลี่ยน state เป็น parent ของมัน
            return path[::-1]                          # กลับลำดับ path ให้เป็น start → goal แล้วส่งกลับ
        zero = state.index(0)                          # หา index ของช่องว่าง (เลข 0) ใน state
        row, col = divmod(zero, 3)                     # แปลง index เป็นตำแหน่งแถว (row) และคอลัมน์ (col)
        for move in (-3, 3, -1, 1):                    # ทดลองขยับช่องว่าง 4 ทิศ: ขึ้น ลง ซ้าย ขวา
            new_row, new_col = divmod(zero + move, 3)  # คำนวณตำแหน่งใหม่หลังจากขยับ
            if 0 <= new_row < 3 and 0 <= new_col < 3 and abs(row - new_row) + abs(col - new_col) == 1:
                # ตรวจสอบว่าอยู่ในขอบเขต 3x3 และเป็นการขยับที่ถูกต้อง (ขยับทีละ 1 ช่อง)
                neighbor = state[:]                    # คัดลอก state เดิม เพื่อสร้างสถานะข้างเคียง
                neighbor[zero], neighbor[zero + move] = neighbor[zero + move], neighbor[zero]
                # สลับตำแหน่งช่องว่าง (0) กับตัวเลขที่ขยับมา
                if tuple(neighbor) not in visited:     # ถ้าสถานะนี้ยังไม่เคยถูกเยี่ยมชม
                    visited[tuple(neighbor)] = state   # บันทึก parent ของ neighbor เป็น state ปัจจุบัน
                    dq.append(neighbor)                # เพิ่ม neighbor ลงคิว BFS

def printSolution(path):                               # ฟังก์ชันสำหรับแสดงเส้นทางคำตอบ
    for state in path:                                 # วนลูปทีละ state ใน path
        print("\n".join(' '.join(map(str, state[i:i+3])) for i in range(0, 9, 3)))
        # แสดง state ในรูปแบบกระดาน 3x3
        print("-----")                                 # เส้นคั่นระหว่างแต่ละ state

# Example Usage
startState = [1, 3, 0, 6, 8, 4, 7, 5, 2]                # กำหนดสถานะเริ่มต้นของบอร์ด

solution = bfs(startState)                             # เรียก BFS เพื่อหาเส้นทางคำตอบ
if solution:                                           # ถ้าพบคำตอบ
    printSolution(solution)                            # แสดงเส้นทางคำตอบทั้งหมด
    print(f"Solved in {len(solution) - 1} moves.")     # แสดงจำนวนครั้งที่ขยับ (จำนวน index state - 1)
else:                                                  # ถ้าไม่พบคำตอบ
    print("No solution found.")                        # แสดงข้อความว่าไม่มีคำตอบ
```

#อธิบายโค้ด โค้ดนี้เป็นการแก้เกม 8-Puzzle โดยใช้วิธี BFS หรือพูดง่าย ๆ คือ ค่อย ๆ ลองทุกทางที่เป็นไปได้แบบเป็นชั้น ๆ ทีละ level ของ tree ไม่รีบลึก แต่เก็บให้ครบก่อน เลยมั่นใจได้ว่าพอเจอคำตอบ จะใช้จำนวนครั้งน้อยที่สุด

* ตรง **deque** เอาไว้ทำคิว เพราะ BFS ต้องหยิบตัวที่เข้าก่อนออกก่อน (popleft() ทำงานเร็วกว่า list) ส่วน visited ใช้เก็บสถานะที่เคยเจอแล้ว เพื่อไม่ให้วนกลับไปเล่นกระดานเดิมซ้ำ ๆ และในขณะเดียวกันก็ใช้จำว่า state นี้มาจาก state ไหน เอาไว้ย้อนเส้นทางตอนเจอคำตอบ


* ในลูปหลัก มันจะดึงกระดานออกมาทีละอัน ถ้ากระดานนั้นเรียงถูกแล้ว (ตรงกับ target) ก็จะเริ่มไล่ย้อนกลับทันที โดยดูจาก visited ว่าแต่ละกระดานมาก่อนหน้านี้คืออะไร แล้วเอามาเรียงเป็น path ตั้งแต่เริ่มจนจบ


* ถ้ายังไม่ถึงคำตอบ โค้ดจะไปหาเลข 0 ก่อน เพราะ 0 คือช่องว่างที่ขยับได้ จากนั้นใช้ divmod แปลงตำแหน่ง index ให้เป็นแถวกับคอลัมน์ จะได้เช็คทิศทางได้ง่ายขึ้น แล้วลองขยับขึ้น ลง ซ้าย ขวา ด้วยค่า -3, +3, -1, +1 ถ้าขยับแล้วไม่หลุดกระดาน และเป็นช่องติดกันจริง ก็ถือว่าเป็น neighbor ที่ใช้ได้


* พอได้กระดานใหม่ ก็ทำสำเนาด้วย state[:] เพื่อไม่ไปแก้ของเดิม แล้วสลับตำแหน่ง 0 กับตัวที่ขยับมา ถ้ากระดานนี้ยังไม่เคยอยู่ใน visited ก็เก็บว่ามันมาจาก state ไหน แล้วโยนเข้าไปในคิว รอให้ BFS ไปจัดการต่อ

* ส่วนฟังก์ชันแสดงผล จะเอากระดานแต่ละขั้นมาแบ่งเป็น 3 แถว แล้วใช้ "-----" คั่นไว้ เพื่อให้เห็นชัด ๆ ว่าแต่ละเส้นคือการขยับหนึ่งครั้ง จบด้วยการบอกว่าทั้งหมดใช้กี่ move

#2 จากโปรแกรมในข้อ 1 ให้นศ. ปรับปรุงโปรแกรมดังกล่าวให้ผู้เล่นสามารถป้อนค่าสถานะของบอร์ดเริ่มต้นได้ และสามารถป้อนค่าสถานะสิ้นสุดของบอร์ดตอนจบเกมได้ โดยมีเงื่อนไขเพิ่มเติมดังนี้

2.1 ถ้าผู้เล่นไม่ป้อนค่าสถานะเริ่มต้น ให้โปรแกรมทำการสุ่มสถานะเริ่มต้น

2.2 ถ้าผู้เล่นไม่ป้อนค่าสถานะสิ้นสุด ให้โปรแกรมใช้ค่าสถานะเป็น
[1, 2, 3, 4, 5, 6, 7, 8, 0] เหมือนเดิม

```python
from collections import deque
import random

def bfs(start_state, target):
    dq = deque([start_state])
    visited = {tuple(start_state): None}

    while dq:
        state = dq.popleft()
        if state == target:
            path = []
            while state:
                path.append(state)
                state = visited[tuple(state)]
            return path[::-1]

        zero = state.index(0)
        row, col = divmod(zero, 3)

        for move in (-3, 3, -1, 1):
            new_row, new_col = divmod(zero + move, 3)
            if 0 <= new_row < 3 and 0 <= new_col < 3 and abs(row - new_row) + abs(col - new_col) == 1:
                neighbor = state[:]
                neighbor[zero], neighbor[zero + move] = neighbor[zero + move], neighbor[zero]
                if tuple(neighbor) not in visited:
                    visited[tuple(neighbor)] = state
                    dq.append(neighbor)


def printSolution(path):
    for state in path:
        print("\n".join(' '.join(map(str, state[i:i+3])) for i in range(0, 9, 3)))
        print("-----")


def get_board_input(prompt):
    user_input = input(prompt).strip()

    if user_input == "":
        return None

    board = list(map(int, user_input.split()))

    if sorted(board) != list(range(9)):
        print("Invalid input! Must contain numbers 0-8 exactly once.")
        return None

    return board


startState = get_board_input(
    "Enter start state (9 numbers 0-8) or press Enter to randomize: "
)

if startState is None:
    startState = list(range(9))
    random.shuffle(startState)


goalState = get_board_input(
    "Enter goal state (9 numbers 0-8) or press Enter for default: "
)

if goalState is None:
    goalState = [1, 2, 3, 4, 5, 6, 7, 8, 0]


solution = bfs(startState, goalState)

if solution:
    printSolution(solution)
    print(f"Solved in {len(solution) - 1} moves.")
else:
    print("No solution found.")
```

```python
from collections import deque          # นำเข้า deque เพื่อใช้เป็นคิว (queue) สำหรับ BFS
import random                           # นำเข้า random เพื่อใช้สุ่มกระดานเริ่มต้น

def bfs(start_state, target):           # ฟังก์ชัน BFS รับกระดานเริ่มต้น และกระดานเป้าหมาย
    dq = deque([start_state])           # สร้างคิว และใส่ state เริ่มต้นเข้าไปเป็นตัวแรก
    visited = {tuple(start_state): None}# เก็บ state ที่เคยเจอ โดย key คือ state และ value คือ state ก่อนหน้า

    while dq:                           # ทำงานตราบใดที่ยังมี state อยู่ในคิว
        state = dq.popleft()            # ดึง state แรกออกจากคิว (FIFO)

        if state == target:             # ถ้า state ปัจจุบันตรงกับเป้าหมาย
            path = []                   # สร้าง list สำหรับเก็บเส้นทางคำตอบ
            while state:                # ไล่ย้อนกลับจาก state เป้าหมายไปยัง state เริ่มต้น
                path.append(state)      # เก็บ state ปัจจุบันลงใน path
                state = visited[tuple(state)]  # ย้อนกลับไปยัง state ก่อนหน้า
            return path[::-1]           # กลับลำดับ path ให้จากต้น → เป้าหมาย

        zero = state.index(0)           # หา index ของช่องว่าง (เลข 0)
        row, col = divmod(zero, 3)      # แปลง index เป็นตำแหน่งแถวและคอลัมน์

        for move in (-3, 3, -1, 1):     # ทดลองขยับช่องว่าง: บน ล่าง ซ้าย ขวา
            new_row, new_col = divmod(zero + move, 3)  # คำนวณตำแหน่งใหม่หลังขยับ

            if 0 <= new_row < 3 and 0 <= new_col < 3 and abs(row - new_row) + abs(col - new_col) == 1:
                neighbor = state[:]     # คัดลอก state เดิมเพื่อสร้าง state ใหม่
                neighbor[zero], neighbor[zero + move] = neighbor[zero + move], neighbor[zero]
                                         # สลับตำแหน่ง 0 กับช่องที่ขยับไป

                if tuple(neighbor) not in visited:     # ถ้ายังไม่เคยเจอ state นี้มาก่อน
                    visited[tuple(neighbor)] = state   # บันทึกว่า state นี้มาจาก state ไหน
                    dq.append(neighbor)                # นำ state ใหม่เข้าคิวเพื่อค้นหาต่อ


def printSolution(path):                                 # ฟังก์ชันแสดงเส้นทางคำตอบ
    for state in path:                                  # วนลูปแสดงทีละ state ใน path
        print("\n".join(' '.join(map(str, state[i:i+3])) for i in range(0, 9, 3)))
                                                         # แสดง state ในรูปแบบกระดาน 3x3
        print("-----")                                  # เส้นคั่นระหว่างแต่ละ state


def get_board_input(prompt):                            # ฟังก์ชันรับกระดานจากผู้ใช้
    user_input = input(prompt).strip()                  # รับ input และลบช่องว่างหัวท้าย

    if user_input == "":                                # ถ้าผู้ใช้กด Enter เปล่า
        return None                                     # ส่ง None กลับไป

    board = list(map(int, user_input.split()))          # แปลง input เป็น list ของตัวเลข

    if sorted(board) != list(range(9)):                 # ตรวจว่ามีเลข 0–8 ครบและไม่ซ้ำ
        print("Invalid input! Must contain numbers 0-8 exactly once.")
        return None                                     # ถ้าผิด ส่ง None

    return board                                        # ถ้าถูกต้อง ส่งกระดานกลับ


startState = get_board_input(
    "Enter start state (9 numbers 0-8) or press Enter to randomize: "
)                                                        # รับกระดานเริ่มต้นจากผู้ใช้

if startState is None:                                  # ถ้าไม่ได้กรอก หรือกรอกผิด
    startState = list(range(9))                         # สร้างกระดานพื้นฐาน
    random.shuffle(startState)                          # สุ่มลำดับให้เป็นกระดานใหม่


goalState = get_board_input(
    "Enter goal state (9 numbers 0-8) or press Enter for default: "
)                                                        # รับกระดานเป้าหมายจากผู้ใช้

if goalState is None:                                   # ถ้าไม่ได้กรอก
    goalState = [1, 2, 3, 4, 5, 6, 7, 8, 0]              # ใช้กระดานเป้าหมายมาตรฐาน


solution = bfs(startState, goalState)                   # เรียก BFS เพื่อหาเส้นทางคำตอบ

if solution:                                            # ถ้ามีคำตอบ
    printSolution(solution)                             # แสดงเส้นทางคำตอบ
    print(f"Solved in {len(solution) - 1} moves.")      # แสดงจำนวนครั้งที่ขยับ
else:
    print("No solution found.")                         # กรณีไม่มีคำตอบ
```

```python
from collections import deque              # นำเข้า deque เพื่อใช้เป็นโครงสร้างคิว (Queue) สำหรับ BFS

def bfs(start_state):                      # ฟังก์ชัน BFS รับสถานะเริ่มต้นของกระดาน
    target = [1,2,3,4,5,6,7,8,0]            # กำหนดสถานะเป้าหมาย (Goal State)
    dq = deque([start_state])               # สร้างคิวและใส่สถานะเริ่มต้นเข้าไป
    visited = {tuple(start_state): None}    # เก็บ state ที่เคยเยี่ยม โดยชี้กลับไปยัง parent

    while dq:                               # ทำงานตราบใดที่คิวยังไม่ว่าง
        state = dq.popleft()                # ดึง state แรกออกจากคิว (FIFO)

        if state == target:                 # ถ้า state ปัจจุบันตรงกับเป้าหมาย
            path = []                       # สร้าง list สำหรับเก็บเส้นทางคำตอบ
            while state:                    # ไล่ย้อนจาก goal ไปยัง start
                path.append(state)          # เก็บ state ปัจจุบันลง path
                state = visited[tuple(state)]  # ย้อนกลับไปยัง parent
            return path[::-1]               # กลับลำดับให้เป็น start → goal

        zero = state.index(0)               # หา index ของช่องว่าง (0)
        row, col = divmod(zero, 3)          # แปลง index เป็นตำแหน่งแถวและคอลัมน์

        for move in (-3, 3, -1, 1):          # ทดลองขยับช่องว่าง: บน ล่าง ซ้าย ขวา
            new = zero + move               # คำนวณตำแหน่งใหม่ของช่องว่าง
            if 0 <= new < 9:                # ตรวจว่าตำแหน่งใหม่อยู่ในกระดาน
                r, c = divmod(new, 3)       # แปลงตำแหน่งใหม่เป็นแถวและคอลัมน์
                if abs(row-r) + abs(col-c) == 1:  # ตรวจว่าขยับติดกันจริง (ไม่ข้าม)
                    nxt = state[:]          # คัดลอก state เดิม
                    nxt[zero], nxt[new] = nxt[new], nxt[zero]  # สลับ 0 กับตำแหน่งใหม่
                    if tuple(nxt) not in visited:  # ถ้ายังไม่เคยเจอ state นี้
                        visited[tuple(nxt)] = state  # บันทึก parent ของ state ใหม่
                        dq.append(nxt)      # ใส่ state ใหม่เข้าคิว


def printLevel(states):                     # ฟังก์ชันแสดง state หลายตัวในระดับเดียวกัน
    for r in range(3):                      # วนทีละแถวของกระดาน (3 แถว)
        for s in states:                   # วนทีละ state ในระดับเดียวกัน
            print(" ".join(map(str, s[r*3:(r+1)*3])), end="     ")
                                            # แสดงแถวเดียวกันของหลาย state ต่อกัน
        print()                             # ขึ้นบรรทัดใหม่เมื่อจบแถว
    print("-----")                          # เส้นคั่นระหว่างแต่ละ level


def showTree(path):                         # ฟังก์ชันแสดง Tree ตามเส้นทางที่ BFS เลือก
    for i in range(len(path)):              # วนทีละระดับของ path
        current = path[i]                  # state ที่ถูกเลือกในระดับนี้
        parent = path[i-1] if i > 0 else None  # parent ของ state ปัจจุบัน
        zero = current.index(0)             # หา index ของช่องว่าง
        row, col = divmod(zero, 3)          # แปลงเป็นตำแหน่งแถว-คอลัมน์
        level = [current]                   # ใส่ state ที่ถูกเลือกเป็นตัวแรกของ level

        if i < len(path)-1:                 # ถ้ายังไม่ใช่ level สุดท้าย
            for move in (-3,3,-1,1):        # ทดลองขยับช่องว่าง
                new = zero + move           # คำนวณตำแหน่งใหม่
                if 0 <= new < 9:            # ตรวจว่าอยู่ในกระดาน
                    r,c = divmod(new,3)     # แปลงเป็นแถว-คอลัมน์
                    if abs(row-r)+abs(col-c)==1:  # ตรวจว่าเป็นการขยับที่ถูกต้อง
                        nb = current[:]     # คัดลอก state ปัจจุบัน
                        nb[zero], nb[new] = nb[new], nb[zero]  # สร้าง neighbor
                        if parent and nb == parent:  # ถ้าเป็น parent เดิม
                            continue        # ข้าม ไม่แสดงซ้ำ
                        level.append(nb)    # เพิ่ม neighbor ที่เป็นไปได้ใน level นี้

        printLevel(level)                   # แสดง state ทั้งหมดใน level นี้


startState = [1, 3, 0, 6, 8, 4, 7, 5, 2]     # กำหนดสถานะเริ่มต้น
solution = bfs(startState)                  # เรียก BFS เพื่อหาเส้นทางคำตอบ

showTree(solution)                          # แสดง Tree ตามเส้นทางที่ BFS เลือก
print(f"Solved in {len(solution) - 1} moves.")  # แสดงจำนวนครั้งที่ขยับ
```

* เริ่มจากฟังก์ชัน printLevel(states) อันนี้มีหน้าที่เอา หลาย ๆ state มาแสดงเรียงกันในระดับเดียวกัน เหมือนเรามอง Tree ทีละชั้น โดยใช้ลูป for r in range(3) เพื่อพิมพ์ทีละแถวของกระดาน 3x3 แล้วเอาแต่ละ state มาวางข้างกันในแนวนอน เวลาพิมพ์ออกมามันเลยดูเหมือน “แตกกิ่ง” ของ Tree จริง ๆ ส่วน "-----" ก็แค่คั่นแต่ละ level ให้ดูง่าย ไม่งงสายตา


* ต่อมาคือ showTree(path) ฟังก์ชันนี้คือหัวใจของข้อ 3 เลย มันเอา path ที่ได้จาก BFS (ซึ่งคือเส้นทางที่ ถูกเลือกจริง ไปจนถึง goal) มาไล่ทีละ step แล้วในแต่ละ step เราจะเอา state ปัจจุบันมาเป็น “โหนดหลัก” จากนั้นลองขยับช่องว่าง 0 ทุกทิศ (บน ล่าง ซ้าย ขวา) เพื่อสร้าง โหนดข้างเคียง ทั้งหมดที่เป็นไปได้ใน level นั้น


* ตรงที่มีตัวแปร parent เอาไว้ ก็เพื่อกันไม่ให้เราพิมพ์โหนดที่ย้อนกลับไปพ่อของมันเอง เพราะอันนั้นไม่ถือว่าเป็นทางเลือกใหม่จริง ๆ ใน Tree เป็นแค่ย้อนกลับเฉย ๆ เลยข้ามไป
