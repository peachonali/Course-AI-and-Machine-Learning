#1. ให้นักศึกษาเขียนโปรแกรมภาษา Python เพื่อคำนวณการทำงานของ Perceptron ในโครงข่ายประสาทเทียม (Neural Networks) โดยครอบคลุมขั้นตอน Feedforward และ Backpropagation พร้อมทั้งแสดงผลลัพธ์การทำงานของโปรแกรมบนหน้าจอ


```python
import math
# Step 0:  เตรียมฟังก์ชันที่ใช้ใน Neural Network (Activation + Derivative)
def sigmoid(x):
    """Sigmoid Activation Function"""
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(output):
    """Derivative of Sigmoid: f'(z) = f(z) * (1 - f(z))  Backpropagation """
    return output * (1 - output)

# Step 1: Feed-forward (คำนวณ Output ของ Neural Network)
def calculate_nn(name, x, w, b, target, eta):
    print(f" --- การคำนวณสำหรับ: {name} --- ")
    # z = (w1*x1 + w2*x2 + ... ) + b
    z = sum(xi * wi for xi, wi in zip(x, w)) + b
    output = sigmoid(z)

    print(f"Step 1 (Feed-forward):")
    print(f" - ค่า Linear Combination (z): {z:.4f}")
    print(f" - ค่า Output (a): {output:.4f}")

# Step 2: Backpropagation (คำนวณ Gradient เพื่ออัปเดต Weight)
    # Find Error (Loss) but this one focus on Gradient for Wi updating
    # dLoss/dOutput - -(target - output)
    error_grad = -(target - output)

    # doutput/dz = output * (1 - output)
    activation_grad = sigmoid_derivative(output)

    print(f"\nStep 2 (Backpropagation):")
    print(f" - Target (y): {target}")
    print(f" - dLoss/da: {error_grad:.4f}")
    print(f" - da/dz: {activation_grad:.4f}")

# Step 3: Weights Update (ปรับค่า Weight ด้วย Gradient Descent)
# Step 3.1: Weights Update
    new_weights = []
    for i in range(len(w)):
        # dz/dwi = xi
        dz_dwi = x[i]
        # gradient รวม = dLoss/da * da/dz * dz/dwi
        gradient = error_grad * activation_grad * dz_dwi
        new_w = w[i] - (eta * gradient)
        new_weights.append(new_w)
        print(f" - Gradient ของ w{i+1}: {gradient:.4f}")
        print(f" - w{i+1} ใหม่: {new_w:.4f}")

# Step 3.2: Bias update
    gradient_b = error_grad * activation_grad
    new_b = b - (eta * gradient_b)
    print(f" - Gradient ของ b: {gradient_b:.4f}")
    print(f" - b ใหม่: {new_b:.4f}")

    return output, new_weights ,new_b

# Step 4:  เรียกใช้งานฟังก์ชัน (ทดสอบด้วยข้อมูลตัวอย่าง)
# --- Data from example in class ---
# Inputs: 0.5, 0.1 | Weights: 0.2, -0.3 | Bias: 0.1 | Target: 1.0 | LR: 0.1
calculate_nn("ตัวอย่างในบทเรียน", [0.5, 0.1], [0.2, -0.3], 0.1, 1.0, 0.1)

print("\n" + "=" * 40 + "\n")

# --- Data from exercise in class ---
# Inputs: 1, 2 | Weights: 0.1, 0.5 | Bias: -0.2 | Target: 0 | LR: 0.5
calculate_nn("แบบฝึกหัด", [1, 2], [0.1, 0.5], -0.2, 0, 0.5)
```

#2. ให้นักศึกษาปรับปรุงโปรแกรมจากข้อที่ 1 .ให้ใช้งานกับชุดข้อมูลตามตัวอย่างในชั้นเรียนเพื่อให้สามารถทำงานกับชุดข้อมูลตัวอย่างที่ใช้ในชั้นเรียน (สไลด์หน้าที่ 10) โดยกำหนดค่าเริ่มต้นดังนี้

Training Data (x1, x2, Label)

dataset = [

    [-0.7, -0.6, 0],
    [-0.6,  0.4, 0],
    [-0.5,  0.7, 1],
    [-0.5, -0.2, 0],
    [-0.1,  0.7, 1],
    [ 0.1, -0.2, 0],
    [ 0.3,  0.5, 1],
    [ 0.4,  0.1, 1],
    [ 0.4, -0.7, 0],
    [ 0.7,  0.2, 1]
    ]

**Initial Parameters**

w1, w2 = 0.5, -0.2   # สุ่มค่าน้ำหนักเริ่มต้น

bias = 0.1           # ค่าไบแอสเริ่มต้น

learning_rate = 0.3  # อัตราการเรียนรู้

epochs = 500         # จำนวนรอบในการเรียนรู้

เมื่อทำการฝึกสอน (Train) ครบ 500 epochs แล้ว ให้นักศึกษาแสดงค่าพารามิเตอร์ล่าสุด ได้แก่ w1, w2 และ bias พร้อมทั้งแสดงผลการทำนาย (Prediction) ของชุดข้อมูลดังกล่าวหลังการ train ด้วย

```python
# Step 0: เตรียม Training Data (input + label)
# Training Data
dataset = [
    [-0.7, -0.6, 0],
    [-0.6,  0.4, 0],
    [-0.5,  0.7, 1],
    [-0.5, -0.2, 0],
    [-0.1,  0.7, 1],
    [ 0.1, -0.2, 0],
    [ 0.3,  0.5, 1],
    [ 0.4,  0.1, 1],
    [ 0.4, -0.7, 0],
    [ 0.7,  0.2, 1]
]

# Step 1: กำหนดค่าเริ่มต้นของโมเดล (Weights + Bias + Hyperparameters)
# Initial Parameters
w1 = 0.5
w2 = -0.2
bias = 0.1
learning_rate = 0.3
epochs = 500

print("--- เริ่มการฝึกสอน ---")
print(f"ค่าเริ่มต้น: w1={w1:.4f}, w2={w2:.4f}, bias={bias:.4f}, Learning Rate={learning_rate}, Epochs={epochs}")

# Step 2: Training Loop (วนเรียนรู้หลายรอบ = Epochs)
# Training Loop
for epoch in range(epochs):

# Step 2.1: วนทุกข้อมูลใน dataset (Stochastic Gradient Descent)
    for x1, x2, target_label in dataset:

# Step 2.2: Feed-forward (คำนวณ output จาก input)
        # z = (w1*x1 + w2*x2) + bias
        z = (w1 * x1) + (w2 * x2) + bias
        output = sigmoid(z)  # sigmoid function from previous cell

# Step 2.3: Backpropagation (คำนวณ gradient ของ loss)
        # dLoss/dOutput = -(target - output)
        error_grad = -(target_label - output)

        # doutput/dz = output * (1 - output)
        activation_grad = sigmoid_derivative(output) # sigmoid_derivative function from previous cell

# Step 2.4: คำนวณ Gradient ของแต่ละพารามิเตอร์
        # dz/dw1 = x1
        gradient_w1 = error_grad * activation_grad * x1
        # dz/dw2 = x2
        gradient_w2 = error_grad * activation_grad * x2
        # dz/dbias = 1
        gradient_bias = error_grad * activation_grad * 1

        # Update weights and bias
        w1 = w1 - (learning_rate * gradient_w1)
        w2 = w2 - (learning_rate * gradient_w2)
        bias = bias - (learning_rate * gradient_bias)

# Step 3: แสดงผลลัพธ์หลังการฝึกสอน
print("\n--- สิ้นสุดการฝึกสอน ---")
print(f"ค่าพารามิเตอร์ล่าสุดหลังการ train {epochs} epochs:")
print(f"w1 = {w1:.4f}")
print(f"w2 = {w2:.4f}")
print(f"bias = {bias:.4f}")

# Step 4: Prediction (ทดสอบโมเดลหลัง train)
print("\n--- ผลการทำนาย (Prediction) ของชุดข้อมูลหลังการ train ---")
print("Input (x1, x2) | Target | Predicted Output (Sigmoid) | Predicted Class")
print("-" * 75)
for x1, x2, target_label in dataset:
    z = (w1 * x1) + (w2 * x2) + bias
    prediction_output = sigmoid(z)
    predicted_class = 1 if prediction_output >= 0.5 else 0
    print(f"({x1:5.2f}, {x2:5.2f}) | {target_label:6d} | {prediction_output:26.4f} | {predicted_class:15d}")
```

#3. ให้นักศึกษาปรับปรุงโปรแกรมจากข้อที่ 2 โดยใช้ Library สำเร็จรูปจาก Scikit-learn เพื่อเปรียบเทียบผลลัพธ์ เช่น w1, w2 และ bias ว่าใกล้เคียงหรือไม่กับผลลัพธ์จากข้อที่ 2 โดยมีการกำหนดให้ค่าพารามิเตอร์ (paremeters) เริ่มต้นดังนี้


model = MLPClassifier(
  
    hidden_layer_sizes=(),
    activation='logistic',
    solver='sgd',
    learning_rate_init=0.3,
    max_iter=500,
    random_state=42  # เพื่อให้ผลลัพธ์คงที่ในการรันแต่ละครั้ง
)

```python
# Step 0: Import library ที่ใช้ (NumPy + Scikit-learn)
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

# Step 1: เตรียมข้อมูล (แยก Input X และ Label y)
  # Prepare data for Scikit-learn
  # Assuming 'dataset' is already defined from previous cells
X = np.array([[d[0], d[1]] for d in dataset])
y = np.array([d[2] for d in dataset])

# Step 2: (Optional) Feature Scaling (ช่วยให้ MLP train ได้ดีขึ้น)
  # Scale the features (important for MLPClassifier performance)
  # scaler = StandardScaler()
  # X_scaled = scaler.fit_transform(X)

# Step 3: สร้างโมเดล MLPClassifier (กำหนดโครงสร้างและ hyperparameters)
# Define the MLPClassifier model as specified
model = MLPClassifier(
    hidden_layer_sizes=(), # No hidden layers for a single-layer perceptron
    activation='logistic', # Corresponds to sigmoid
    solver='sgd',          # Stochastic Gradient Descent
    learning_rate_init=0.3,
    max_iter=500,
    random_state=42,       # For reproducible results
    # verbose=True, # Uncomment to see training progress
    tol=1e-4,              # Tolerance for optimization
    n_iter_no_change=10    # Number of iterations with no improvement to wait before stopping
)

# Step 4: Training (ฝึกโมเดลด้วยข้อมูล X, y)
print("\n--- เริ่มการฝึกสอนด้วย Scikit-learn ---")
model.fit(X, y)
print("--- สิ้นสุดการฝึกสอนด้วย Scikit-learn ---")

print("\n--- ค่าพารามิเตอร์ล่าสุดจาก Scikit-learn ---")
# MLPClassifier stores weights in model.coefs_ and biases in model.intercepts_
# For a single-layer perceptron (no hidden layers), coefs_ will have one element for weights
# and intercepts_ will have one element for bias.

# Step 5: ดึงค่า Weight และ Bias จากโมเดลที่ train แล้ว
w1_sklearn = model.coefs_[0][0][0] # First feature's weight
w2_sklearn = model.coefs_[0][1][0] # Second feature's weight
bias_sklearn = model.intercepts_[0][0] # Bias for the output layer

print(f"w1 (Scikit-learn) = {w1_sklearn:.4f}")
print(f"w2 (Scikit-learn) = {w2_sklearn:.4f}")
print(f"bias (Scikit-learn) = {bias_sklearn:.4f}")

# Step 6: Prediction (ทำนายผลหลังการ train)
print("\n--- ผลการทำนาย (Prediction) ของชุดข้อมูลหลังการ train ด้วย Scikit-learn ---")
print("Input (x1, x2) | Target | Predicted Probability | Predicted Class")
print("-" * 75)

predictions_proba = model.predict_proba(X)
predictions_class = model.predict(X)

# Step 6.1: แสดงผลลัพธ์ทีละข้อมูล
for i, (x1, x2, target_label) in enumerate(dataset):
    predicted_prob = predictions_proba[i][1] # Probability of class 1
    predicted_class = predictions_class[i]
    print(f"({x1:5.2f}, {x2:5.2f}) | {target_label:6d} | {predicted_prob:21.4f} | {predicted_class:15d}")
```
