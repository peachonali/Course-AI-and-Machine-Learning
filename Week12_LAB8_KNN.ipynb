# 1. ให้นักศึกษาเขียนโปรแกรมภาษา Python เพื่อแก้ปัญหา Classification ด้วย k-Nearest Neighbor โดยกำหนดให้มีชุดข้อมูล Training Data ต่อไปนี้

training_data = [

{'Point': 'P1', 'X1': 2, 'X2': 4, 'Class': 'A'},

{'Point': 'P2', 'X1': 4, 'X2': 2, 'Class': 'A'},

{'Point': 'P3', 'X1': 3, 'X2': 5, 'Class': 'A'},

{'Point': 'P4', 'X1': 7, 'X2': 8, 'Class': 'B'},

{'Point': 'P5', 'X1': 8, 'X2': 7, 'Class': 'B'},

{'Point': 'P6', 'X1': 6, 'X2': 9, 'Class': 'B'},

{'Point': 'P7', 'X1': 5, 'X2': 3, 'Class': 'C'},

{'Point': 'P8', 'X1': 6, 'X2': 4, 'Class': 'C'},
]


**กำหนดให้โปรแกรมจะต้องมีความสามารถต่อไปนี้**


• ผู้ใช้งานสามารถป้อนค่า k ได้ โดยมีค่าเริ่มต้นเป็น k = 3

• ผู้ใช้งานสามารถป้อนค่า Test Data ได้ โดยมีค่าเริ่มต้นเป็น X1 = 5 และ X2 = 6

• แสดงค่า Euclidean Distance จาก Test Data ไปยัง Training Data ทุกตัว

• แสดงค่า Distance ที่น้อยที่สุด k ตัว และแสดงให้เห็นว่าแต่ละตัวอยู่ใน Class ใด

• แสดงค่าการทำนาย (Prediction) ของ Test
Data ว่าอยู่ใน Class ใด

```python
import math

# Step 1: เตรียม Training Data
# ================================
training_data = [
    {'Point': 'P1', 'X1': 2, 'X2': 4, 'Class': 'A'},
    {'Point': 'P2', 'X1': 4, 'X2': 2, 'Class': 'A'},
    {'Point': 'P3', 'X1': 3, 'X2': 5, 'Class': 'A'},
    {'Point': 'P4', 'X1': 7, 'X2': 8, 'Class': 'B'},
    {'Point': 'P5', 'X1': 8, 'X2': 7, 'Class': 'B'},
    {'Point': 'P6', 'X1': 6, 'X2': 9, 'Class': 'B'},
    {'Point': 'P7', 'X1': 5, 'X2': 3, 'Class': 'C'},
    {'Point': 'P8', 'X1': 6, 'X2': 4, 'Class': 'C'}
]

# Step 2: ฟังก์ชันแสดงตารางข้อมูล
# ================================
def show_training_data(data):
    print("===== 1. ตาราง Training Data =====")
    print("Point\tX1\tX2\tClass")
    for d in data:
        print(f"{d['Point']}\t{d['X1']}\t{d['X2']}\t{d['Class']}")

# Step 3: รับค่า Input
# ================================
def get_input():
    k_input = input("\nใส่ค่า k (default = 3): ")
    k = int(k_input) if k_input else 3

    x1_input = input("ใส่ X1 (default = 5): ")
    x2_input = input("ใส่ X2 (default = 6): ")

    x1 = float(x1_input) if x1_input else 5
    x2 = float(x2_input) if x2_input else 6

    print("\n===== 2. ค่าที่รับจากผู้ใช้ =====")
    print(f"k = {k}")
    print(f"Test Data = ({x1}, {x2})")

    return k, x1, x2

# Step 4: คำนวณ Euclidean Distance
# ================================
def calculate_distances(data, x1, x2):
    results = []

    print("\n===== 3. ตาราง Euclidean Distance =====")
    print("Point\tX1\tX2\tClass\tDistance")

    for d in data:
        dx = x1 - d['X1']
        dy = x2 - d['X2']
        dist = math.sqrt(dx * dx + dy * dy)

        row = {
            'Point': d['Point'],
            'X1': d['X1'],
            'X2': d['X2'],
            'Class': d['Class'],
            'Distance': dist
        }

        results.append(row)

        print(f"{row['Point']}\t{row['X1']}\t{row['X2']}\t{row['Class']}\t{dist:.4f}")

    return results


# Step 5: เลือก K ที่ใกล้ที่สุด
# ================================
def get_k_nearest(distances, k):
    sorted_data = sorted(distances, key=lambda x: x['Distance'])

    print("\n===== 4. K ตัวที่ใกล้ที่สุด =====")
    print("Point\tClass\tDistance")

    k_nearest = sorted_data[:k]

    for d in k_nearest:
        print(f"{d['Point']}\t{d['Class']}\t{d['Distance']:.4f}")

    return k_nearest

# Step 6: นับคะแนนและทำนายผล
# ================================
def predict_class(k_nearest):
    class_count = {'A': 0, 'B': 0, 'C': 0}

    for d in k_nearest:
        class_count[d['Class']] += 1

    print("\n===== 5. การนับคะแนน =====")
    for key in class_count:
        print(f"Class {key} = {class_count[key]}")

    prediction = max(class_count, key=class_count.get)

    print("\n===== ผลการทำนาย =====")
    print(f"Test Data อยู่ใน Class {prediction}")

# Step 7: Main Program
# ================================
def main():
    show_training_data(training_data)
    k, x1, x2 = get_input()
    distances = calculate_distances(training_data, x1, x2)
    k_nearest = get_k_nearest(distances, k)
    predict_class(k_nearest)


# เรียกใช้งานโปรแกรม
main()
```

# 2. ให้นักศึกษาเขียนโปรแกรมภาษา Python เพื่อแก้ปัญหา Clustering ด้วย k-Mean Clustering โดยมีชุดข้อมูลต่อไปนี้

POINTS = {
'A': (1, 1),

'B': (2, 3),

'C': (3, 2),

'D': (4, 5),

'E': (6, 5),

'F': (7, 7),

'G': (8, 6)
}

**โดยกำหนดให้ค่าเริ่มต้นมีดังนี้**

• จำนวนกลุ่ม k = 2

• จุด Centroid ของกลุ่มที่ 1 = (2, 2) และจุด Centroid ของกลุ่มที่ 2 = (7, 6)

**จากนั้นให้โปรแกรมแสดงข้อมูลการคำนวณในแต่ละรอบทุกรอบ จนกระทั่งหยุดการทำงานดังนี้**

• ตารางการคำนวณและ Assign แต่ละ Point ให้ Centroid ที่ใกล้ที่สุด

• การคำนวณหาค่า Centroid ใหม่ของแต่ละกลุ่ม

```python
import math

# Step 1: เตรียมข้อมูล
# ================================
POINTS = {
    'A': (1, 1),
    'B': (2, 3),
    'C': (3, 2),
    'D': (4, 5),
    'E': (6, 5),
    'F': (7, 7),
    'G': (8, 6)
}

# ค่าเริ่มต้น
k = 2
centroids = [(2, 2), (7, 6)]

# Step 2: แสดงตารางข้อมูล
# ================================
def show_points(points):
    print("===== 1. ตารางข้อมูล =====")
    print("Point\tX\tY")
    for name, (x, y) in points.items():
        print(f"{name}\t{x}\t{y}")

# Step 3: คำนวณ Distance + Assign
# ================================
def assign_points(points, centroids):
    clusters = {i: [] for i in range(len(centroids))}

    print("\n===== 2. ตารางคำนวณ Distance =====")
    print("Point\tC1\tC2")

    assignment_table = []

    for name, (x, y) in points.items():
        dists = []

        for cx, cy in centroids:
            dx = x - cx
            dy = y - cy
            dist = math.sqrt(dx*dx + dy*dy)
            dists.append(dist)

        print(f"{name}\t{dists[0]:.4f}\t{dists[1]:.4f}")

        # หา centroid ที่ใกล้ที่สุด
        min_index = dists.index(min(dists))
        clusters[min_index].append((name, x, y))

        assignment_table.append((name, dists, min_index))

    return clusters, assignment_table

# Step 4: แสดงผลการ Assign
# ================================
def show_assignment(assignment_table):
    print("\n===== 3. การ Assign กลุ่ม =====")
    print("Point\tCluster")

    for name, dists, idx in assignment_table:
        print(f"{name}\tC{idx+1}")

# Step 5: คำนวณ Centroid ใหม่
# ================================
def update_centroids(clusters):
    new_centroids = []

    print("\n===== 4. คำนวณ Centroid ใหม่ =====")

    for i in clusters:
        points = clusters[i]

        sum_x = 0
        sum_y = 0

        for name, x, y in points:
            sum_x += x
            sum_y += y

        if len(points) > 0:
            new_x = sum_x / len(points)
            new_y = sum_y / len(points)
        else:
            new_x, new_y = 0, 0

        new_centroids.append((new_x, new_y))

        print(f"C{i+1} ใหม่ = ({new_x:.2f}, {new_y:.2f})")

    return new_centroids

# Step 6: นับจำนวนจุดในแต่ละกลุ่ม
# ================================
def count_points(clusters):
    print("\n===== 5. จำนวนสมาชิกในแต่ละกลุ่ม =====")
    for i in clusters:
        print(f"Cluster C{i+1} =", len(clusters[i]), "points")

# Step 7: Main Program (วนจนกว่าจะหยุด)
# ================================
def main():
    show_points(POINTS)

    current_centroids = centroids
    iteration = 1

    while True:
        print(f"\n========== Iteration {iteration} ==========")

        clusters, assignment_table = assign_points(POINTS, current_centroids)
        show_assignment(assignment_table)
        count_points(clusters)

        new_centroids = update_centroids(clusters)

        # เช็คหยุด (centroid ไม่เปลี่ยน)
        if new_centroids == current_centroids:
            print("\n*** หยุดการทำงาน (Centroid ไม่เปลี่ยนแล้ว) ***")
            break

        current_centroids = new_centroids
        iteration += 1


# เรียกใช้งานโปรแกรม
main()
```

# 3. ให้นักศึกษาเขียนโปรแกรมใหม่โดยใช้ Library สำเร็จรูปจาก Scikit Learn เพื่อเปรียบเทียบกับโปรแกรมที่นักศึกษาเขียนมาก่อนหน้านี้

• เปรียบเทียบผลลัพธ์ k-NN จากข้อ 1 ว่าเหมือนหรือแตกต่างกันอย่างไร เมื่อใช้ Scikit Learn

• เปรียบเทียบผลลัพธ์ k-Mean จากข้อ 2 ว่าเหมือนหรือแตกต่างกันอย่างไร เมื่อใช้ Scikit Learn

```python
# Step 1: เตรียมข้อมูล
# ================================
from sklearn.neighbors import KNeighborsClassifier

X = [
    [2, 4],
    [4, 2],
    [3, 5],
    [7, 8],
    [8, 7],
    [6, 9],
    [5, 3],
    [6, 4]
]

y = ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C']

# Step 2: กำหนดค่า k และ Test Data
# ================================
k = 3
test_data = [[5, 6]]

print("===== k-NN (Scikit-Learn) =====")
print("k =", k)
print("Test Data =", test_data[0])

# Step 3: สร้างโมเดลและ Train
# ================================
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X, y)

# Step 4: ทำนายผล
# ================================
prediction = model.predict(test_data)

print("\nPrediction =", prediction[0])
```

```python
# Step 1: เตรียมข้อมูล
# ================================
from sklearn.cluster import KMeans

X = [
    [1, 1],
    [2, 3],
    [3, 2],
    [4, 5],
    [6, 5],
    [7, 7],
    [8, 6]
]

# Step 2: กำหนดค่าเริ่มต้น
# ================================
initial_centroids = [[2, 2], [7, 6]]

print("\n===== k-Means (Scikit-Learn) =====")

# Step 3: สร้างโมเดล
# ================================
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1)

# Step 4: Train โมเดล
# ================================
kmeans.fit(X)

# Step 5: แสดงผลลัพธ์
# ================================
print("\nCluster ของแต่ละจุด:")
for i, label in enumerate(kmeans.labels_):
    print(f"Point {X[i]} -> Cluster {label+1}")

print("\nScikit-learn Centroids:")
print("\nCentroid สุดท้าย:")
for i, c in enumerate(kmeans.cluster_centers_):
    print(f"C{i+1} = ({c[0]:.2f}, {c[1]:.2f})")
```
