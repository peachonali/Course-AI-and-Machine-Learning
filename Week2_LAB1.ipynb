#Part 1 Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
# Input Graph
graph = {
    'A': ['B' , 'C' , 'Y'],
    'B': ['A' , 'C' , 'D' , 'X'],
    'C': ['A' , 'B' , 'E'],
    'D': ['B' , 'E'],
    'E': ['C' , 'D'],
    'X': ['B'],
    'Y': ['A']
    # ถ้าเริ่ม C จะแวะ [C] > [A B E] > [ B E ] > [ E 'D'] > [ D ]  แวะแค่ตัวแรก C A B D E
    # เริ่ม ฺB จะแวะ[B] > [A C D] > [C D] > [C D E] > [D E] > [E]

}   # การเขียนกราฟ Node ที่เชื่อมต้องเชื่อมต่อทุก Node ด้วย
# Create Graph object
G = nx.Graph()
# Add edge from dictionary
for node in graph:
    for neighbor in graph[node]:
      G.add_edge(node,neighbor)

# Define position of nodes
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(6,6))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2000,
    node_color='lightblue',
    edge_color='gray',
    font_size=10,
    font_color='black',
    font_weight='bold',
    width=2,
)
plt.title("Graph Representation Searching Algorithm" , fontsize=16)
plt.show()
```

#Part 2 Breadth-First Search BFS

```python
 from ast import Add
##### Breadth-First Search (BFS) #####
# To store visited nodes.
visitedNodes = []
# To store nodes in queue
queueNodes = [ ]
# function
def bfs(visitedNodes, graph, snode):
##### ให้นักศึกษาเขียนโปรแกรมตาม Pseudocode ที่กำหนดในใบงาน #####
  visitedNodes.append(snode)
  queueNodes.append(snode)

  while queueNodes:
    node = queueNodes.pop(0) # Get the first node from the queue
    print(node, end=" ") # Print the visited node
    for neighbor in graph[node]:
      if neighbor not in visitedNodes:
        visitedNodes.append(neighbor)
        queueNodes.append(neighbor)

# Main Code
snode = input("Enter Starting Node(A, B, C, D, E, or X) : ").upper()
# calling bfs function
print("\nRESULT :")
bfs(visitedNodes, graph, snode)
```

#Part 3 Depth-First Search DFS

```python
##### Depth-First Search (DFS) #####
# Set used to store visited nodes.
visitedNodes = []
# function
def dfs(visitedNodes, graph, node):
##### ให้นักศึกษาเขียนโปรแกรมตาม Pseudocode ที่กำหนดในใบงาน #####
  if node not in visitedNodes:
    print(node, end=" ")
    visitedNodes.append(node)
    for neighbor in graph[node]:
      dfs(visitedNodes, graph, neighbor)


# Main Code
snode = input("Enter Starting Node(A, B, C, D, E, or X) : ").upper()
# calling dfs function
print("\nRESULT :")
dfs(visitedNodes, graph, snode)
```
